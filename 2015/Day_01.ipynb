```julia
open("data\\day_01.txt", "r") do file
    global lines = readlines(file)
end
data = lines[1]
```

# Puzzle 1

```julia
# l -> length of string
# u -> number of '(' == up
# d -> number of ')' == down
# l = u + d
# floor = u - d = u - (l - u) = 2 * u - l

l = length(data)
u = count(c -> (c == '('), data)
d = count(c -> (c == ')'), data)

println(u - d)
println(2*u - l)
```

# Puzzle 2

```julia
steps = [c == '(' ? 1 : -1 for c in data];
floor = cumsum(steps)
first_below_0 = findfirst(floor .< 0)
println(first_below_0)

plot(floor)
vline!([first_below_0])
annotate!([(first_below_0, minimum(floor), string(first_below_0))])
```
