```julia
using DelimitedFiles
using Plots
gr()
theme(:dark)
```

```julia
depths = readdlm("data\\day_01.txt", '\t', Int, '\n');
size(depths)
```

# Puzzle 1

```julia
changes = depths[2:end] - depths[1:end-1];
size(changes)
```

```julia
count(i->(i>0), changes)
```

```julia
plot([-depths, -changes], t=[:line :scatter], title=["Height" "Changes"],
     layout=2, size=[800, 300])
```

# Puzzle 2

```julia
windows = (depths[1:end-2] + depths[2:end-1] + depths[3:end])/3
size(windows)
```

```julia
changes = windows[2:end] - windows[1:end-1];
size(changes)
```

```julia
count(i->(i>0), changes)
```

```julia
plot([-depths, -windows, -changes], t=[:line :line :scatter], title=["Height" "Height (3-point average)" "Changes"],
     layout=(1, 3), size=[1000, 300])
```
