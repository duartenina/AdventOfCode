```julia
using DataFrames
using CSV
```

```julia
filename = "data\\day_04.txt"
```

```julia
open(filename, "r") do file
   global lucky_numbers = parse.(Int, split(readline(file), ','))
end
@show lucky_numbers
@show size(lucky_numbers)
;
```

```julia
raw_2d_data = Matrix(CSV.read(filename, DataFrame, skipto=2, ignoreemptyrows=true, delim=' ', ignorerepeated=true, header=false))
```

```julia
data_3d = reshape(raw_2d_data, (5, :, 5));
for i in 1:3
    @show data_3d[:, i, :]
end
```

# Puzzle 1

```julia
temp_data = copy(data_3d)
high_value = 10_000
for num in lucky_numbers
    @show num
    temp_data[temp_data .== num] .= high_value
    
    rowsum = sum(temp_data, dims=1)
    colsum = sum(temp_data, dims=3)
    
    rowfull = findall(rowsum .== (high_value * 5))
    colfull = findall(colsum .== (high_value * 5))
    
    square_num = -1
    
    if length(rowfull) > 0
        @show rowfull
        square_num = rowfull[1][2]
    end
    
    if length(colfull) > 0
        @show colfull
        square_num = colfull[1][2]
    end
    
    if square_num > -1
        @show square_num
        square = temp_data[:, square_num, :]
        @show square
        @show sum(square[square .!= high_value]) * num
        break
    end
end
```

# Puzzle 2

```julia
temp_data = copy(data_3d)
high_value = 10_000
for num in lucky_numbers
    @show num
    temp_data[temp_data .== num] .= high_value
    
    rowsum = sum(temp_data, dims=1)
    colsum = sum(temp_data, dims=3)
    
    rowfull = findall(rowsum .== (high_value * 5))
    colfull = findall(colsum .== (high_value * 5))
    
    square_nums = vcat(
        [i[2] for i in rowfull],
        [i[2] for i in colfull]
    )
    
    if size(temp_data)[2] < 2
        if length(square_nums) > 0
            square_num = square_nums[1]
        else
            square_num = 1
        end
        square = temp_data[:, square_num, :]
        @show square
        @show sum(square[square .!= high_value]) * num
        
        if length(square_nums) > 0
            break
        end
    end
    
    temp_data = temp_data[:, (!in).(1:end, Ref(square_nums)), :]
    @show size(temp_data)
end
```
