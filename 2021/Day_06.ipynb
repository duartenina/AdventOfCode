```julia
using DelimitedFiles, Statistics
using Plots
using LsqFit
gr()
theme(:dark)
```

```julia
starting_ages = readdlm("data\\day_06.txt", ',', Int, '\n')
```

# Puzzle 1

```julia
num_days = 80
ages = copy(starting_ages)
days = 0:num_days
counts = [length(ages)]
for i in 1:num_days
    ages .-= 1
    births = ages .== -1
    num_births = count(births)
    
    ages[births] .= 6
    ages = hcat(ages, ones(Int, 1, num_births) * 8)
    append!(counts, length(ages))
    #@show ages
end
@show length(ages);
```

```julia
plot(days, counts)
```

# Puzzle 2

```julia
age_counts = [count(starting_ages .== i) for i in 0:8]
num_days = 80

for day in 1:num_days
    births = age_counts[1]
    age_counts[1:end-1] .= age_counts[2:end]
    age_counts[7] += births
    age_counts[end] = births
end
@show age_counts
@show sum(age_counts)
```
